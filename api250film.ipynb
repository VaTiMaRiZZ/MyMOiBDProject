```python
import requests
import pandas as pd
```

```python
TOKEN = 'ASW159B-1RX4B52-HJ2HQAE-JPB0MCA' 
API_URL = 'https://api.poiskkino.dev/v1.4/movie'
FILE_NAME = '250ruFilm.csv' 

headers = {'X-API-KEY': TOKEN}

params = {
    'limit': 250,
    'type': 'movie',
    'countries.name': ['!Россия', '!СССР'],
    'sortField': 'rating.kp',
    'sortType': '-1',
    'votes.kp': '100000-10000000',
    'selectFields': ['name', 'alternativeName', 'description', 'year', 'countries', 'genres']
}

response = requests.get(API_URL, headers=headers, params=params)

if response.status_code == 200:
    api_data = response.json().get('docs', [])
    
    foreign_movies = []
    for m in api_data:
        foreign_movies.append({
            'name': m.get('name') or m.get('alternativeName'),
            'description': m.get('description') or 'Нет описания',
            'country': ', '.join([c['name'] for c in m.get('countries', [])]),
            'year': m.get('year'),
            'genre': ', '.join([g['name'] for g in m.get('genres', [])])
        })

    df_foreign = pd.DataFrame(foreign_movies)
    df_russian = pd.read_csv(FILE_NAME)
    df_final = pd.concat([df_russian, df_foreign], ignore_index=True)
    df_final.to_csv(FILE_NAME, index=False, encoding='utf-8-sig')
    print(f"Успех")
        
else:
    print(f"Ошибка {response.status_code}")
```
